```python
from torch.nn import CrossEntropyLoss, Sigmoid, ReLU, Softsign
from TrainingEnvironments.FeedForwardTrainingEnvironment import FeedForwardTrainingEnvironment
from Datasets.GaussianSample import GaussianData

from Models.MLP import MLP
import torch.optim as optim 
import matplotlib.pyplot as plt
```

```python
# Experiment 1
input_shape = 10
output_shape = 1028
data = GaussianData(output_shape, input_shape)
data.showData()
```

# Experiment 1

```python
hidden_shapes = [10, 10, 10, 10]
activation_function = ReLU

model = MLP(input_shape, hidden_shapes, output_shape, activation_function)
criterion = CrossEntropyLoss()
optimizer = optim.SGD(model.parameters(), lr=0.1)

epochs = 50
batches = 100
batchsize = 10

Experiment1 = FeedForwardTrainingEnvironment(data, model, criterion, optimizer, epochs, batches, batchsize)
loss_data, accuracy_data, mean_activation_data, std_activation_data = Experiment1.trainModel()
```

```python
# Plot loss and accuracy curves
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(loss_data, label='Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Loss Curve')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(accuracy_data, label='Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.title('Accuracy Curve')
plt.legend()

plt.show()
```

```python
# Plotting the mean values with error bars representing the standard deviation
plt.figure(figsize=(10, 6))
plt.errorbar(range(epochs), mean_activation_data[0], yerr=std_activation_data[0], fmt='-o', capsize=5, capthick=1, color='red', ecolor='red')
plt.errorbar(range(epochs), mean_activation_data[1], yerr=std_activation_data[1], fmt='-o', capsize=5, capthick=1, color='green', ecolor='green')
plt.errorbar(range(epochs), mean_activation_data[2], yerr=std_activation_data[2], fmt='-o', capsize=5, capthick=1, color='blue', ecolor='blue')
plt.errorbar(range(epochs), mean_activation_data[3], yerr=std_activation_data[3], fmt='-o', capsize=5, capthick=1, color='black', ecolor='black')

# Labeling the plot
plt.xlabel('Epoch')
plt.ylabel('Mean Value')
plt.title('Mean Values Across Epochs with Standard Deviation')
plt.legend(['Layer1', 'Layer2', 'Layer3', 'Layer4'])
plt.grid(True)

# Display the plot
plt.savefig('Mean Activation Values Across Epochs with Standard Deviation.png')
plt.show()
```
